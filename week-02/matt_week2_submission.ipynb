```python
from MyCapytain.resources.texts.local.capitains.cts import CapitainsCtsText

with open("../tei/tlg0525.tlg001.perseus-grc2.xml") as f:
    text = CapitainsCtsText(urn="urn:cts:greekLit:tlg0525.tlg001.perseus-grc2", resource=f)
```

```python
from lxml import etree
from MyCapytain.common.constants import Mimetypes

urns = []
raw_xmls = []
unannotated_strings = []

for ref in text.getReffs(level=len(text.citation)):
    urn = f"{text.urn}:{ref}"
    node = text.getTextualNode(ref)
    raw_xml = node.export(Mimetypes.XML.TEI)
    tree = node.export(Mimetypes.PYTHON.ETREE)
    s = etree.tostring(tree, encoding="unicode", method="text")

    urns.append(urn)
    raw_xmls.append(raw_xml)
    unannotated_strings.append(s)
```

```python
%pip install numpy==1.26.4

%pip install pandas
```

```python
import pandas as pd

d = {
    "urn": pd.Series(urns, dtype="string"),
    "raw_xml": raw_xmls,
    "unannotated_strings": pd.Series(unannotated_strings, dtype="string")
}
pausanias_df = pd.DataFrame(d)
```

```python
import spacy

nlp = spacy.load("grc_proiel_sm", disable=["ner"])
```

```python
tokenizer = nlp.tokenizer

pausanias_df['tokens'] = pausanias_df['unannotated_strings'].apply(tokenizer)
```

```python
pausanias_df['whitespaced_tokens'] = pausanias_df['unannotated_strings'].str.split()

pausanias_df
```

```python
pausanias_df['whitespaced_tokens'].explode().count()
```

```python
len(pausanias_df['whitespaced_tokens'].explode().unique())
```

```python
from collections import Counter

type_counts = Counter(pausanias_df['whitespaced_tokens'].explode())

type_counts.most_common(100)
```

```python
types = [t.text for t in pausanias_df['tokens'].explode() if not t.is_stop and t.is_alpha]

type_counts = Counter(types)

type_counts.most_common(100)
```

```python
raw_texts = [t for t in pausanias_df['unannotated_strings']]
annotated_texts = nlp.pipe(raw_texts, batch_size=100)

pausanias_df['nlp_docs'] = list(annotated_texts)
```

```python
lemmata = [t.lemma_ for t in pausanias_df['nlp_docs'].explode() if not t.is_stop and t.is_alpha]

lemmata_counts = Counter(lemmata)

lemmata_counts.most_common(100)
```

Analysis/Commentary: 

Right away, I noticed that the first three Books of Pausanias differ quite a bit in terms of length -- i.e., token count. Book 1 has the greatest token count at 51,265 tokens, which is significantly more than Book 2's 22,835 tokens, which is in turn greater than Book 3's 17,737 tokens. I find it interesting that Book 1 is so much larger than than Books 2 and 3; I am not yet very familiar with  Pausanias' work, so perhaps this difference in token count would make sense in light of the content of the Books. 

I think the data regarding the most common lemmata in each Books is interesting. Many of the same words appear in all three, especially 'ἐνταῦθα.' Given that the work is a description of various places throughout ancient Greece, it's not surprising that ἐνταῦθα would be mentioned so many times. I think this data has the potential to be extremely helpful to students, as it gives us a sense of which vocabularly is most important to learn in order to best understand and read the text. However, some lemmata are prevalent only in certain books, which may give us insight into the book's contents. For example, the lemma 'Ἀθηναῖος' and its forms appear 161 times in Book 1; 'Ἀργεῖος' and its forms appear 99 times in Book 2; and 'Λακεδαιμόνιος' and its forms appear 125 times in Book 3. Based on this data, it seems that each Book may focus primarily on a different area of Greece, with Book 1 describing Athens, Book 2 Argos, and Book 3 Sparta. I am curious to see if this turns out to be true once I've read the text. 

I have chosen to take a closer look at Book 3, in English translation for ease of reading. I expected to see many variations of terms related to the Lacedaemonians (Spartans), and that largely panned out. Throughout the entire third Book, terms such as Lacedaemon (both referring to a person and later to the city-state) and Lacedaemonians were exceedingly common; in ancient Greek these would all be variations of the lemma 'Λακεδαιμόνιος'. I was correct in my hunch that Book 3 would focus on Sparta, and if other books in the Periegesis follow a similar format, I bet they each focus on a different part of Greece as I suspected.  

Although Book 3 was about Lacedaemon/Sparta as I hypothesized, the actual content of the chapter was not quite what I had expected. Based on what little I knew about the Periegesis, I was expecting more of Book 3 to describe the city-state and its surroundings more. While it certainly does do this, it isn't really until 3.11, when Pausanias describes the market-place in Sparta; he then goes on to describe other parts of the city-state as well as its monuments. The first ten chapters of Book 3 instead ended up being largely dedicated to describing the mytho-history of the Lacedaemonians. I expected some of this, but thought that it would have been smaller than it ended up being. Even when he was describing the city-state and its monuments, he included much commentary on their cultural/mytho-history -- again more than I was expecting. Before taking a closer look at Book 3, I thought more emphasis would be placed on eyewitness accounts of monuments, locations, geography, etc. -- the physical aspects of ancient Greece -- but I realized that he focused much more on the culture and history. This was a pleasant surprise, as I found the commentary on the mytho-history throughout more interesting than the physical descriptions. 

My main take-away from my analysis is that although a quick lemmata count and analysis can yield helpful insights, the actual content may differ -- or be presented differently -- from what is expected. But perhaps I should have expected more emphasis on the cultural history of the people in the first place given that the lemma was 'Λακεδαιμόνιος' instead of 'Λακεδαίμων.'
